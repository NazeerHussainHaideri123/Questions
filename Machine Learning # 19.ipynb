```python
# Use scikit-learn's PCA to visualize the principal components of a high-dimensional dataset.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=100, n_features=20, centers=3, random_state=42)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
n_components = 2  # Number of components for visualization
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(X_scaled)
```

```python
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(pc_df['PC1'], pc_df['PC2'], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - Principal Components Visualization')
plt.colorbar(label='Class')
plt.show()
```
