```python
# Implement Hierarchical Clustering on a sample dataset and create a dendrogram to visualize the hierarchical structure.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=20, n_features=2, centers=3, random_state=42)
```

```python
linked = linkage(X, method='ward')  # You can change the linkage method as needed
```

```python
plt.figure(figsize=(10, 6))
dendrogram(linked, orientation='top', labels=y, distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
